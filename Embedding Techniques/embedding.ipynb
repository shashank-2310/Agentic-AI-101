### Embedding Techniques
**Converting text into vectors** is a fundamental step in many natural language processing tasks. This process, known as embedding, allows us to represent words, sentences, or even entire documents as numerical vectors that capture their semantic meaning. In this notebook, we will explore various embedding techniques and how to implement them using Python.

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
os.environ["LANGSMITH_API_KEY"]=os.getenv("LANGSMITH_API_KEY") or ""
os.environ["LANGSMITH_TRACING"]=os.getenv("LANGSMITH_TRACING") or ""
os.environ["LANGSMITH_PROJECT"]=os.getenv("LANGSMITH_PROJECT") or ""
```

#### Using OpenRouter's embedding method with OpenAI's embedding model

```python
from openrouter import OpenRouter

with OpenRouter(api_key=OPENROUTER_API_KEY) as open_router:
    query_result = open_router.embeddings.generate(
        input="This is a tutorial on Embedding Techniques.",
        model="text-embedding-3-large",
    )

print(query_result)
```

```python
print(type(query_result))
print(type(query_result.data[0].embedding)) # pyright: ignore[reportAttributeAccessIssue]
print(len(query_result.data[0].embedding))  # pyright: ignore[reportAttributeAccessIssue]
```

##### Using Custom Dimensions

```python
from openrouter import OpenRouter

with OpenRouter(api_key=OPENROUTER_API_KEY) as open_router:
    query_result = open_router.embeddings.generate(
        input="This is a tutorial on Embedding Techniques.",
        model="text-embedding-3-large",
        dimensions=1024,  # Specify custom dimensions
    )

query_result
```

#### Using OpenAI's embedding method & model with OpenRouter's API key and base URL

```python
from langchain_openai import OpenAIEmbeddings
from pydantic import SecretStr

if OPENROUTER_API_KEY is None:
    raise ValueError("OPENROUTER_API_KEY is missing. Set it in your .env file.")

embeddings_1024=OpenAIEmbeddings(
    api_key=SecretStr(OPENROUTER_API_KEY),
    base_url="https://openrouter.ai/api/v1",
    model="text-embedding-3-large",
    dimensions=1024 #Custom dimension for the embedding vector
)

embeddings_1024
```

```python
type(embeddings_1024)
```

```python
text = "This is a tutorial on Embedding Techniques"
query_result = embeddings_1024.embed_query(text)

print(len(query_result))
print(query_result)
```

### Embedding for Doc

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
final_doc = text_splitter.split_documents(docs)
final_doc
```

#### Vector Embedding & Vector StoreDB
Using vector embeddings in conjunction with a vector database allows us to efficiently store and retrieve high-dimensional data, such as text or images, based on their semantic similarity.  
This is particularly useful for tasks like information retrieval, recommendation systems, and natural language processing.  
In this section, we will explore how to create vector embeddings and store them in a vector database for efficient querying and retrieval.  
Chromadb will be used as the vector database here, and it is a popular choice for handling large-scale vector data due to its efficiency and scalability.

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_doc, embeddings_1024)
db
```

```python
# Retrieve the most similar document to the query

query="It will be all the easier for us to conduct ourselves as belligerents"
res = db.similarity_search(query)   # Similarity search

print(type(res))
print(res)
```
