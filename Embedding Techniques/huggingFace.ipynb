### HuggingFace Embedding Techniques
In this notebook, we will explore how to use Hugging Face's embedding techniques to convert text into numerical vectors that can be used for various natural language processing tasks. We will use the `transformers` library to access pre-trained models and generate embeddings for our text data.

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['HF_TOKEN'] = os.getenv("HUGGINGFACE_TOKEN")
```

#### Sentence Transformers on Hugging Face
Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class. We have also added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "The Eiffel Tower is located in Paris."
res = embeddings.embed_query(text)
res
```

```python
len(res)
```

```python
doc_res = embeddings.embed_documents([text, "The Great Wall of China is located in China."])
doc_res[0]
```
