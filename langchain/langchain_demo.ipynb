LangChain is an open-source framework used to build applications powered by large language models (LLMs) like OpenAI GPT, Llama, or Anthropic Claude.

It helps developers connect LLMs with data, tools, APIs, and workflows to build real AI apps instead of just sending prompts to a model.

Think of it like “middleware for AI apps.”

Why LangChain Exists?

A raw LLM alone can only answer based on its training data.
- But real applications need things like:
- Access documents / databases
- Call APIs
- Remember conversation history
- Perform multi-step reasoning

LangChain provides the building blocks to do that.

```python
import os
from urllib import response

from langchain.agents import create_agent
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from pydantic import SecretStr

load_dotenv()
```

```python
llm = ChatGroq(
    model=os.getenv("MODEL") or "llama-3.1-8b-instant",
    api_key=SecretStr(os.getenv("GROQ_API_KEY") or "")
)

agent = create_agent (
    model=llm,
    system_prompt="You're an expert in GenAI and can explain it in simple terms to a 5 year old.",
)
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "Can you tell me about Langsmith?"}]}
)

print(response)
```

```python
type(response)
```

```python
print(response["messages"][-1].content)
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()

response = agent.invoke(
    {"messages": [{"role": "user", "content": "Can you tell me about Langsmith?"}]}
)

parsed_response = output_parser.parse(str(response["messages"][-1].content))
print(parsed_response)
```
