LangChain is an open-source framework used to build applications powered by large language models (LLMs) like OpenAI GPT, Llama, or Anthropic Claude.

It helps developers connect LLMs with data, tools, APIs, and workflows to build real AI apps instead of just sending prompts to a model.

Think of it like “middleware for AI apps.”

Why LangChain Exists?

A raw LLM alone can only answer based on its training data.
- But real applications need things like:
- Access documents / databases
- Call APIs
- Remember conversation history
- Perform multi-step reasoning

LangChain provides the building blocks to do that.

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENROUTER_API_KEY']=os.getenv("OPENROUTER_API_KEY") or ""
## Langsmith Tracking
os.environ["LANGSMITH_API_KEY"]=os.getenv("LANGSMITH_API_KEY") or ""
os.environ["LANGSMITH_TRACING"]=os.getenv("LANGSMITH_TRACING") or ""
os.environ["LANGSMITH_PROJECT"]=os.getenv("LANGSMITH_PROJECT") or ""
```

```python
# from langchain_openai import ChatOpenAI
from langchain_openrouter import ChatOpenRouter
from openai import max_retries

llm=ChatOpenRouter(
    model=os.getenv("MODEL") or "openai/gpt-oss-120b:free", 
    temperature=0.8, 
    max_retries=2
)
print(llm)
```

```python
## Input and get response form LLM

result=llm.invoke("What is generative AI?")

print(result)
```

```python
### Chatprompt Template
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI Engineer. Provide me answers based on the questions"),
        ("user","{input}")
    ]

)
prompt
```

```python
## chain 
chain=prompt|llm

response=chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response)
```

```python
## stroutput Parser

from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser

response=chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response)
```
