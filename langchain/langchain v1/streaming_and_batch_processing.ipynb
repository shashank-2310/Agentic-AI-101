### Streaming and Batch

#### Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are
produced. You can use a loop to process each chunk in real-time.  


```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["OPENROUTER_API_KEY"]=os.getenv("OPENROUTER_API_KEY", "")
os.environ["LANGSMITH_API_KEY"]=os.getenv("LANGSMITH_API_KEY", "")
os.environ["LANGSMITH_TRACING"]=os.getenv("LANGSMITH_TRACING", "")
os.environ["LANGSMITH_PROJECT"]=os.getenv("LANGSMITH_PROJECT", "")
```

```python
from langchain_openrouter import ChatOpenRouter

model = ChatOpenRouter(model = "gpt-oss-20b:free")
model
```

```python
model.invoke("Why do parrots have colorful feathers?")
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="", flush=True)
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="|", flush=True)
```

##### Invoke vs Stream
invoke() returns the complete output after the model has finished processing, while stream() provides output in chunks as it is generated.  
Streaming allows for real-time interaction and faster feedback, especially for longer responses, while invoke() is suitable for cases where you want the entire response at once.

### Batch
Batching is a collection of independent requests to a model that are processed together. Batching can significantly improve performance and reduce costs as the parallel processing is done.


```python
response = model.batch([
    "What is the capital of France?",
    "What is the largest mammal?",
    "What is the meaning of life?"
])
for res in response:
    print(res.content)
```

```python
response = model.batch(
    [
        "What is the capital of France?",
        "What is the capital of Germany?",
        "How many continents are there on Earth?",
        "Which language is the most spoken worldwide?"
    ],
    config = {
        'max_concurrency': 5 # Limit to number of parallel requests
    }
)
for res in response:
    print(res.content)
```
